1. Email Validator

```python
email = input("Enter email: ")

print(email)

has_at = False
for ch in email:
    if ch == "@":
        has_at = True

ends_ok = False
if email.endswith(".com") or email.endswith(".org"):
    ends_ok = True

if has_at and ends_ok:
    print("Valid Email")
else:
    print("Invalid Email")
```

2. Count Word Frequency in Customer Reviews

```python
text = input("Enter review: ").lower()
print(text)
words = text.split()

freq = {}

for w in words:
    w2 = ""
    for ch in w:
        if ch.isalnum():
            w2 += ch
    if w2 == "":
        continue

    if w2 in freq:
        freq[w2] += 1
    else:
        freq[w2] = 1

print(freq)
```

3. Password Strength Checker

```python
pw = input("Enter password: ")


has_upper = has_lower = has_digit = has_special = False
special = "!@#$%^&*()-_+=[]{}|;:'\",.<>?/`~"

for ch in pw:
    if ch.isupper():
        has_upper = True
    elif ch.islower():
        has_lower = True
    elif ch.isdigit():
        has_digit = True
    elif ch in special:
        has_special = True

if has_upper and has_lower and has_digit and has_special:
    print("Strong Password")
else:
    print("Weak Password")
```

4. Username Generator

```python
full = input("Enter full name: ").split()

first = full[0]
last = full[-1]

first3 = first[:3].lower()
last3 = last[-3:].lower()

num = 0
for ch in first + last:
    num += 1  

username = first3 + last3 + str(num)
print(username)
```

5. SMS Word Counter


```python
msg = input("Enter message: ")

words = len(msg.split())
chars = len(msg)

print("Words =", words)
print("Characters =", chars)

if chars > 160:
    print("Status = Exceeds Limit")
else:
    print("Status = Within Limit")
```

6. Remove Duplicates from Contact List

```python
data = input("Enter names: ").split(",")
clean = []


for name in data:
    n = name.strip()
    if n not in clean:
        clean.append(n)

print(", ".join(clean))
```

7. Detect Suspicious Words in Text

```python
text = input("Enter text: ").lower()
banned = ["spam", "scam", "fake"]

warning = False

for i in banned:
    if i in text:
        warning = True

if warning:
    print("Warning: Suspicious content detected")
else:
    print("Clean")
```

9. Count Hashtags and Mentions

```python
post = input("Enter post: ")

hashtags = mentions = 0

for ch in post:
    if ch == "#":
        hashtags += 1
    elif ch == "@":
        mentions += 1

print("Hashtags =", hashtags)
print("Mentions =", mentions)
```

10. Word Reversal for Data Encryption

```python
sentence = input("Enter sentence: ").split()
result = []

for word in sentence:
    rev = ""
    for i in range(len(word)-1, -1, -1):
        rev += word[i]
    result.append(rev)

print(" ".join(result))

```
