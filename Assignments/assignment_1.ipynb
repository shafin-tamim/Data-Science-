Assignment 1: Temperature Check

```python
temperature = float(input("Enter the current temperature : "))

if temperature < 10:
    print("It's too cold!")
elif 10 <= temperature <= 25:
    print("It's cool outside")
else:
    print("It's hot outside!")
```

Assignment 2: ATM Withdrawal System

```python
balance = float(input("Enter your account balance: "))
withdraw = float(input("Enter the amount you want to withdraw: "))

if withdraw > balance:
    print("Insufficient Balance")
else:
    remaining = balance - withdraw
    print("Transaction Successful")
    print("Remaining Balance:", remaining)
```

Assignment 3: Student Grade Calculator

```python
marks = float(input("Enter your marks (0â€“100): "))

if 90 <= marks <= 100:
    print("Grade: A+")
elif 80 <= marks <= 89:
    print("Grade: A")
elif 70 <= marks <= 79:
    print("Grade: B")
elif 60 <= marks <= 69:
    print("Grade: C")
elif 0 <= marks < 60:
    print("Grade: Fail")
else:
    print("Invalid input! Marks must be between 0 and 100.")
```

Assignment 4: Bus Ticket Discount

```python
age = int(input("Enter your age: "))

base_price = 100  

if age < 5:
    final_price = 0
    print("Ticket is Free")
elif 5 <= age <= 18:
    final_price = base_price * 0.5
    print("50% Discount Applied")
elif age >= 60:
    final_price = base_price * 0.7
    print("30% Discount Applied")
else:
    final_price = base_price
    print("Full Price Ticket")

print("Ticket Price:", final_price)
```

Assignment 5: Restaurant Bill with Discount

```python
bill = float(input("Enter total bill amount: "))

if bill > 1000:
    discount = bill * 0.10
    final_bill = bill - discount
    print("10% discount applied!")
else:
    final_bill = bill
    print("No discount applied.")

print("Final bill amount:", final_bill)
```

Assignment 7: Traffic Light System

```python
color = input("Enter a color (red, yellow, or green): ").lower()
if color == "red":
    print("Stop")
elif color == "yellow":
    print("Ready to go")
elif color == "green":
    print("Go")
else:
    print("Invalid color")
```

Assignment 8: Leap Year Checker

```python
year = int(input("Enter a year: "))
if year % 400 == 0:
    print("Leap Year")
elif year % 4 == 0 and year % 100 != 0:
    print("Leap Year")
else:
    print("Not a Leap Year")
```
